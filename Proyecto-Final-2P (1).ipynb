## Descripción del fenómeno

### Movimiento browniano
El movimiento browniano es el movimiento aleatorio que se observa en las partículas que se hallan en un medio fluido (líquido o gas), como resultado de choques contra las moléculas de dicho fluido.1
Este fenómeno de transporte recibe su nombre en honor al escocés Robert Brown, biólogo y botánico. En 1827, mientras miraba a través de un microscopio a las partículas atrapadas en las cavidades dentro de un grano de polen en el agua, señaló que las partículas se movían a través del líquido; pero no fue capaz de determinar los mecanismos que provocaron este movimiento.

### Descripción del tiempo

De más general, el moviemiento browniano modela una caminata aleatoria en tiempo continuo, donde una particula evoluciona en el espacio dando pasos aleatorios independientes en todas las direcciones. Matemáticamente, el movimiento browniano es un proceso estocástico continuo particular de Markov.

### Acciones

En el mercado bursátil se realizan a diario miles de operaciones de compra y venta de acciones, en los cuales intervienen una gran variedad de precios. Por lo cual tenemos dos opciones los compradores y los vendedores. Nuestra parte aleatoria son los rendimientos de la accion.
El precio de una acción no es sinónimo de su valor.

Lo que determina definitivamente la cotización de unas acciones es la oferta y demanda en el mercado. Una empresa emite una cantidad limitada de títulos, por lo que si hay más compradores que vendedores, subirá su precio.

La oferta y la demanda dependen de la valoración que los inversores hacen sobre la empresa que emite las acciones. Los principales factores de los que depende esta valoración son las expectativas sobre el beneficio futuro de la sociedad, su tasa de crecimiento y la evolución prevista de los tipos de interés. Y otros factores más generales también influyen, como las expectativas sobre la evolución económica y la confianza de los inversores.

 

La forma en que se generan los precios se basa en las normas estructuradas por las bolsas, para que durante las sesiones no se sobrepasen unos límites de variación máxima y mínima. Cuando estas sesiones finalizan obtendremos dos precios de referencia: el precio de cierre y el precio medio ponderado. Las fases que se siguen para llegar a dichos precios son principalmente tres:
        

1.-Se negocia el precio de apertura de la acción para la sesión del día respetando los límites máximos que se establezcan (llamado rango estático).
2.-Se introduce la acción en el mercado y se cruzan las ordenes de compra y venta. Si se supera el límite máximo del rango dinámico se subasta un nuevo precio de referencia. La oferta y la demanda marcarán el devenir del precio durante la sesión.
3.-Se cierra la sesión y se establece el precio de cierre que se usará como referente para negociar el precio de la jornada siguiente.

## Características del modelo

### El Movimiento Browniano Generalizado en acciones:
    
En esta sección, se asume que el precio de las acciones que no pagan dividendos  siguen un Movimiento Browniano Geométrico (MBG). En este modelo, se asume que los precios están gobernados por una media y una volatilidad constante. Si S es el precio de una acción en el periodo de tiempo t, la tasa de drift corresponde a µS, donde µ  es el rendimiento esperado de la acción. Esto significa que en un intervalo de tiempo relativamente pequeño como ∆t, y en ausencia de riesgo, la variación esperada en el precio de la acción S es equivalente a µS∆t. Veamos, en ausencia de termino aleatorio (∆B = 0), el cambio en los precios ∆S puede ser escrito como: 

    ∆S = µS∆t

    Cuando ∆t → 0, la ecuación anterior puede reescribirse como: 

        dS = µSdt


        Fácilmente puede encontrase que

        ST = S0e µT

        Esta última ecuación, implica que en ausencia de riesgo o volatilidad, el precio de la acción S, debería crecer a una tasa constante igual a µ. Sin embargo, en el mundo real, los inversionistas enfrentan riesgos de múltiples índoles, políticos, económicos y sociales, entre otros, que derivan en volatilidad. De esta manera, debe incluirse el componente estocástico dado por el diferencial del movimiento Browniano dBt . El MBG puede describirse formalmente como:

            dS = µSdt + σSdB
 
Puede demostrase fácilmente que, 
ST = S0e µ∆t+σ √ ∆t

### Primeras simulaciones



```python
import random
import matplotlib.pylab as plt
import numpy as np
 
 
def random_walk(n):
        x = 0
        c = [x]
        for i in range(n):
               prob = random.random()
               if prob > 0.5:
                       x = x + 1
               elif prob < 0.5:
                       x = x - 1
               c.append(x)
        return c
```

```python
numberOfSteps = 100
walk = random_walk(numberOfSteps)
 
n = np.arange(numberOfSteps+1)
 
```

```python
plt.plot(n, walk, 'r-')
plt.plot(n, walk, 'bo')
plt.xlabel('Time (n)')
plt.ylabel('Position (x)')
plt.show()
```
